```python
import os
import pandas as pd
```

```python
root = os.getcwd()

data_path = os.path.join(root, "final")
files = [x for x in os.listdir(data_path) if "cleaned_" in x]
files
```

```python
cols_to_append = ["OPENADMET_LOGAC50", "OPENADMET_LOGAC50_STD"]

# Load the data
df_ahr = pd.read_parquet(os.path.join(data_path, 'cleaned_AHR_tox21_data.parquet'))
df_pxr = pd.read_parquet(os.path.join(data_path, 'cleaned_PXR_tox21_data.parquet'))

# Rename columns to include protein suffix
df_ahr = df_ahr.rename(columns={col: f"{col}_AHR" for col in cols_to_append})
df_pxr = df_pxr.rename(columns={col: f"{col}_PXR" for col in cols_to_append})

combined = pd.concat([df_ahr, df_pxr], ignore_index=True)
combined
```

```python
combined.to_parquet(os.path.join(data_path, "combined_AHR_PXR_data.parquet"), index=False)
```
